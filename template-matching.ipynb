## Template Matching

1. Template Matching is the simplest form of object detection
2. It simply scans a larger image for a provided template by sliding the template target image across the larger image

```python
# libs
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
main_image = cv2.imread('main_img.jpg')
main_image = cv2.cvtColor(main_image,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(main_image)
```

```python
sample_image = cv2.imread('sample.jpg')
sample_image = cv2.cvtColor(sample_image,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(sample_image)
```

```python
print("Sample Image:",sample_image.shape)
print("Main Image:",main_image.shape)
```

```python
methods = ['cv2.TM_CCOEFF','cv2.TM_CCOEFF_NORMED','cv2.TM_CCORR','cv2.TM_CCORR_NORMED','cv2.TM_SQDIFF','cv2.TM_SQDIFF_NORMED']
```

```python
for m in methods:
    # create copy of image
    main_copy = main_image.copy()
    method = eval(m)
    
    res = cv2.matchTemplate(main_copy,sample_image,method)
    
    min_val,max_val,min_loc,max_loc = cv2.minMaxLoc(res)
    
    if method in [cv2.TM_SQDIFF,cv2.TM_SQDIFF_NORMED]:
        top_left = min_loc
    else:
        top_left = max_loc
    
    height,width,channel = sample_image.shape
    
    bottom_right=(top_left[0]+width,top_left[1]+height)
    cv2.rectangle(main_copy,top_left,bottom_right,(0,0,255),10)
    
    plt.subplot(121)
    plt.imshow(res)
    plt.title("Heatmap template matching")
    plt.subplot(122)
    plt.imshow(main_copy)
    plt.title("Detected")
    
    plt.suptitle(m)
    
    plt.show()
    print('\n')
    print('\n')
```
