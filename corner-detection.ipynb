### Corner Detection

- Harris Corner Detection
- Shi-Tomashi corner Detection

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
flat_chess = cv2.imread('flat-chess.jpg')
flat_chess = cv2.cvtColor(flat_chess,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(flat_chess)
```

```python
gray_flat_chess = cv2.cvtColor(flat_chess,cv2.COLOR_RGB2GRAY)
```

```python
plt.imshow(gray_flat_chess,cmap='gray')
```

```python
real_chess = cv2.imread('real-chess-board.jpg')
real_chess = cv2.cvtColor(real_chess,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(real_chess)
```

```python
gray_real_chess = cv2.cvtColor(real_chess,cv2.COLOR_RGB2GRAY)
plt.imshow(gray_real_chess,cmap='gray')
```

### Harris Corner Detection Algorithm

```python
gray = np.float32(gray_flat_chess)
```

```python
dst = cv2.cornerHarris(src = gray,blockSize = 2, ksize = 3, k = 0.04)
```

```python
dst = cv2.dilate(dst,None)
```

```python
flat_chess[dst>0.01*dst.max()]= [0,0,0]
```

```python
plt.imshow(flat_chess)
```

```python
# for real chess board
gray_c = np.float32(gray_real_chess)
```

```python
d = cv2.cornerHarris(src = gray_c,blockSize =2, ksize = 3, k = 0.04)
```

```python
d = cv2.dilate(d,None)
```

```python
real_chess[d>0.01*d.max()] = [255,0,0]
```

```python
plt.imshow(real_chess)
```
