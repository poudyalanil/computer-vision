## Edge Detection

- Canny Edge Detection

1. Applying Gaussian Blur
2. Intention Gradient
3. Non-Maximum Supression
4. Double Threshold
5. Track Edge by hysteresis

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
image = cv2.imread("sample.jpg")
image = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(image)
```

```python
# double threshold is used to find out potential edges
edges = cv2.Canny(image = image,threshold1 = 127, threshold2 =127 )
```

```python
plt.imshow(edges)
```

```python
med_val = np.median(image)
med_val
```

```python
# lower threshold to either 0 or 70% of the median value wichever is greater
lower = int(max(0,0.7*med_val))
# upper threshold to either 13% of tee median or max value(255) whichever is smaller
upper = int(min(255,1.3*med_val))
```

```python
edges = cv2.Canny(image = image,threshold1 = lower, threshold2 =upper )
plt.imshow(edges)
```

```python
blurred_img = cv2.blur(image,ksize = (5,5))
```

```python
edges = cv2.Canny(image = blurred_img,threshold1 = lower, threshold2 =upper )
plt.imshow(edges)
```
